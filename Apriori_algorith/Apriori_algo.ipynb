```python
import itertools
from itertools import chain, combinations
from collections import defaultdict
import pandas as pd
```

```python
df1=("store_data.csv")

def dataFile(fname):
    data=[]
    df=open(df1,'rU')
    for line in df:
        line=line.strip().rstrip(',')
        #record = (line.split(','))
        record1=frozenset(line.split(','))
        data.append(record1)
    return data 
```

## count function


```python
def count(item_set,n):
    freq_set=defaultdict(int)
    for i in data:
        for j in item_set:
            if len(j)>=n:
                if set(j).issubset(i):
                    freq_set[j]+=1
    return freq_set   
```

## function for candidate set 

```python
def candidateset(itemset,n):
    
   #data=dataFile("store_data.csv") 
   combinations = list() 
   if n==1:
        Cset = defaultdict(int) 
        for i in itemset:
           for j in i:
               Cset[j]+=1
        return Cset       
   if n==2:
      
  # if depth==2:
       for i in range(1,n+1):
           combinations += [i for i in itertools.combinations(itemset,i)]
       #return combinations 
    
       ss=count(combinations,n)
       return ss
   else:
       for i in itemset:
       # print(set(i))
           for j in itemset:
               intersect={a for a in i if a in j}
               if len(intersect)==n-2:
                    #print(intersect)
                    #print(set(i).union(set(j)))
                    ss=tuple(set(i).union(set(j)))
                    combinations.append(ss)
       ss=count(combinations,n)
       return ss           
```

## function for frequent item set

```python
def freq(itemset,mini_support):
    
    data=dataFile("store_data.csv") 
    freqSet=defaultdict(int)
    for i in itemset.keys():
       #print(i)   
       if itemset[i]> mini_support:
           freqSet[i]+=0
    return freqSet
```

## Minimum Support Count

```python
def mini_support(percentage,len_data):
    s=(percentage*len_data)/100
    return s
```
